### Exercise 2 - Turn the shopping cart program into an object-oriented program

```python
# Create a class called cart that retains items and has methods to add, remove, and show

class Cart:
    def __init__(self):
        self.items = []
    
    def add(self, product, qty):
        for item in self.items:
            if item[0].name == product.name:
                self.items.remove(item)
                self.items.append((product, qty))
                return
        self.items.append((product, qty))
            
    def remove(self, product):
        for item in self.items:
            if item[0].name == product:
                self.items.remove(item)
                break
                
    def show(self):
        print("")
        print("==============================================")
        print("                   Receipt")
        print("==============================================")
        for item in self.items:
            print(f"Product: {item[0].name}\nQty: {item[1]}\nUnit Price: ${item[0].price}")
            print("===================")
        print(f"Total: ${self.total()}")
        print("")
    
    def total(self):
        total = 0
        for item in self.items:
            total += item[0].price * item[1]
        return round(total,2)

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

p1 = Product("Apple", .89)
p2 = Product("Pear", .59)
p3 = Product("Banana", .19)

def showMenuHeader():
    print("")
    print("==============================================")
    print("                  Main Menu")
    print("==============================================")
    return

def showIntro():
    print("")
    print("           ************************")
    print("           *  Shopping Cart 1.01  *")
    print("           ************************")
    return

def main():
    showIntro()
    basket = Cart()
    while True:
        showMenuHeader()
        print("")
        print("Enter the number of your choice:")
        print("1) Add/Update Item")
        print("2) Delete Item")
        print("3) Display Receipt")
        print("0) Quit")
        x = input("Menu #: ")
        if x.isdigit():
            x = int(x)
        if x == 1:
            addEntry(basket)
        elif x == 2:
            delEntry(basket)
        elif x == 3:
            receipt(basket)
        elif x == 0:
            receipt(basket,True)
            break

def addEntry(basket):
    print("")
    print("==============================================")
    print("               Add/Update Item")
    print("==============================================")
    while True:
        print("")
        print("Enter the number of your choice:")
        print("1) Apple")
        print("2) Pear")
        print("3) Banana")
        print("0) Return to Main Menu")
        x = input("Menu #: ")
        if x.isdigit():
            x = int(x)
        if x == 1:
            y = input("Total Quantity? : ")
            try:
                if y > 0:
                    basket.add(p1, int(y))
                else:
                    print("Qty Error: Adding 1 item")
                    basket.add(p1, 1)
            except:
                print("Qty Error: Adding 1 item")
                basket.add(p1, 1)
            return
        elif x == 2:
            y = input("Total Quantity? : ")
            try:
                if y > 0:
                    basket.add(p2, int(y))
                else:
                    print("Qty Error: Adding 1 item")
                    basket.add(p2, 1)
            except:
                print("Qty Error: Adding 1 item")
                basket.add(p2, 1)
            return
        elif x == 3:
            y = input("Total Quantity? : ")
            try:
                if y > 0:
                    basket.add(p3, int(y))
                else:
                    print("Qty Error: Adding 1 item")
                    basket.add(p3, 1)
            except:
                print("Qty Error: Adding 1 item")
                basket.add(p3, 1)
            return
        elif x == 0:
            return

def delEntry(basket):
    
    if preCheck(basket) == False: return
    
    print("")
    print("==============================================")
    print("                 Delete Item")
    print("==============================================")
    while True:
        print("")
        print("Enter the number of your choice:")
        print("1) Apple")
        print("2) Pear")
        print("3) Banana")
        print("0) Return to Main Menu")
        x = input("Menu #: ")
        if x.isdigit():
            x = int(x)
        if x == 1:
            basket.remove("Apple")
            print("Removed Apple(s)")
            return
        elif x == 2:
            basket.remove("Pear")
            print("Removed Pear(s)")
            return
        elif x == 3:
            basket.remove("Banana")
            print("Removed Banana(s)")
            return
        elif x == 0:
            return

def receipt(basket,noWait=False):
    if preCheck(basket,noWait) == False: return
    basket.show()
    if noWait == False: waitforuser()
    return

def preCheck(basket,noWait=False):
    if basket.total() == 0:
        print("")
        print("==============================================")
        print("               No Entries")
        print("==============================================")
        if noWait == False: waitforuser()
        return False
    else:
        return True

def waitforuser():
    print("")
    input("         Press Enter to continue              ")
    return

main()
```
